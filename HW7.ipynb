```python
# W7A1

def binary_search(arr,target) :
    left , right = 0 , len(arr) - 1
    while left <= right :
        mid = (left + right) // 2 
        if  arr[mid] == target :
            return mid 
        elif arr[mid] < target :
            left  =  mid + 1 
        else :
            right = mid - 1
    return -1   

arr = list(map(int,input().split()))
target = int(input())

print (binary_search(arr,target))
```

```python
# W7A2

def count_occurrences(arr, x) :
    a = 0 
    for i in arr :
        if i == x :
            a += 1
    return a 
    

arr = list(map(int,input().split()))
x = int(input())

print(count_occurrences(arr,x))
```

```python
# W7A3

def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        # swapped = False
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j] 
                # swapped = True
        # if not swapped :
    return arr

arr = list(map(int,input().split()))

print(bubble_sort(arr))
```

```python
# W7A4

def search(arr) :
    d = {}
    for i in arr :
        if i not in d :
            d[i] = 1
        else :
            d[i] += 1
    max_k = 'a'
    max_v = 0
    for k,v in d.items() :
        if max_v < v :
            max_k , max_v = k ,v 
    return f'{max_k} xuat hien nhieu nhat , som nhat , {max_v} lan'

arr = list(map(int,input().split()))

print(search(arr))
    
```

```python
# W7A8
def nhap_ds(a) :    
    my_ds = ''
    ds_final = []
    for _ in a :
        if _ not in ['[[',']]',' ',']','['] :
            my_ds += _
    my_ds = list(map(int,my_ds.split(',')))
    for i in range(0,len(my_ds)-1,2) :
        ds_final.append([my_ds[i],my_ds[i+1]])
    return ds_final

def nhap(a) :
    a_str = ''
    for _ in a :
        if _ not in ['[',']'] :
            a_str += _
    return list(map(int,a_str.split(',')))

interval = nhap_ds(input())
queries = nhap(input())

# print(interval,queries)
kq = []
for q in queries :
    time = []
    for i in interval :
        if i[0] <= q <= i[1] :
            time.append(i[1] - i[0] +1)

    if time == [] :
        kq.append(-1)
    else :
        kq.append(min(time))

print(kq)
```

```python
# W7A10

a = input()
m = int(input())

def nhap(xau) :
    a_str = ''
    for _ in xau :
        if _ not in ['[',']'] :
            a_str += _
    return list(map(int,a_str.split(',')))
nums = nhap(a)

A = []

for i in range(1 , len(nums) + 1) :
    A.append(sum(nums[0:i]) % m)
# print(A)

kq = max(A)

for i in range(len(A)-1) :
    for j in range(i + 1 , len(A)) :
        if (A[j] - A[i]) >= 0 and (A[j] - A[i]) > kq :
            kq = A [j] - A[i]
        elif (A[j] - A[i]) < 0  and (A[j] - A[i] + m) > kq :
            kq = A [j] - A[i] + m
    
print(kq)
```
