```python
# W5A3

def is_prime(n) :
    if n < 2 :
        return False
    if n == 2 :
        return True
    for i in range(2,int(n**(1/2))+1) :
        if n % i == 0 :
            return False 
    return True

is_prime(int(input()))
```

```python
# W5A4

def p_number(n) :
    a = 0
    for i in range(1,n) :
        if n % i == 0 :
            a += i
    if a == n : return True
    return False 

p_number(int(input()))
```

```python
# W5A5

A = list(map(int,input().split()))
n = int(input())

def search(n,my_list) :
    for i in range(len(my_list)) :
        if my_list[i] == n :
            return i + 1
    return -1

search(n,A)
```

```python
# W5A6

def giai_thua(n) :
    if n == 0 or n ==1 :
        return 1
    else :
        return n*giai_thua(n-1)
    
giai_thua(int(input()))
```

```python
# W5A7

def may_tinh(a,operat,b) :
    a,b=map(float,[a,b])
    if operat == '+' : return f'{a+b:.2f}'
    elif operat == '-' : return f'{a-b:.2f}'
    elif operat == '/' : return f'{a/b:.2f}'
    elif operat == '*' : return f'{a*b:.2f}'

a,operat,b = input().split()
may_tinh(a,operat,b)
```

```python
# W5A8

def hamming(a,b) :
    l_a = [1]
    l_b = [1]

    while a // 2 != 0 :
        l_a.insert(1,a%2)
        a //= 2
    while b // 2 != 0 :
        l_b.insert(1,b%2)
        b //= 2

    if len(l_a) < len(l_b) :
        for i in range(len(l_b)-len(l_a)) :
            l_a.insert(0,0)
    if len(l_b) < len(l_a) :
        for i in range(len(l_a)-len(l_b)) :
            l_b.insert(0,0)
    
    kcach = 0
    for i in range(len(l_b)) :
        if l_a[i] != l_b[i]:
            kcach += 1
    return kcach

    
a,b = map(int,input().split())

hamming(a,b)

```

```python
# W5A9

def tong_chu_so(n) :
    n = str(n)
    tong = 0
    for i in n :
        tong += int(i)
    return tong

tong_chu_so(int(input()))
```

```python
# W5A10

def check(a,b) :
    for i in range(len(a)):
        for j in range(len(a)):
            if a[i] == a[j] and b[i] != b[j]: return False
            elif a[i] != a[j] and b[i] == b[j]: return False
    return True

a,b = input().split()

check(a,b)
```
