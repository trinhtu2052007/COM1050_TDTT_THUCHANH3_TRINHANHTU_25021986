```python
# W8A3
import math
class Check :
    def __init__ (self ,x,y ) :
        self.x = x
        self.y = y
    def ktra_M (self) :
        if self.x == 0 and self.y == 0 :
            return "là gốc toạ độ"
        elif self.x == 0 and self.y != 0 : return "nằm trên trục Oy"
        elif self.x != 0 and self.y == 0 : return "nằm trên trục Ox"
        else : return 'không nằm trên trục , không là gốc toạ độ '
    def kcach(self) :
        return f'{math.sqrt((self.x)**2 + (self.y)**2) :.2f}'
    

x,y = list(map(float,input().split()))

p = Check(x,y)
print(p.ktra_M())
print(p.kcach())

```

```python
# W8A5

class ShoppingCart() :
    def __init__(self) :
        self.gio_hang = []
    def them_sp(self) :
        sp = {}
        ten_sp , gia = input().split()
        sp[ten_sp] = int (gia)
        return (self.gio_hang).append(sp)
    def xoa_sp(self) :
        ten = input()
        self.gio_hang = [item for item in self.gio_hang if ten not in item]
    def ktra(self) :
        a = len(self.gio_hang)
        if a == 0 : print('Rỗng')
        else : print('Không rỗng')
    def tong_tien(self) :
        tong = 0
        for i in self.gio_hang :
            for key in i :
                tong += i[key]
        print('Tổng giá = ',tong)
    def hien_thi(self) :
        a = len(self.gio_hang)
        if a == 0 : print('Không có sản phẩm nào !')
        else :
            for j in self.gio_hang :
                print(j)
    def xoa(self) :
        (self.gio_hang).clear()
    def menu(self) :
        k = int(input('Nhập lựa chọn :\n1.Thêm vào giỏ hàng một sản phẩm mới\n2. Xóa khỏi giỏ một sản phẩm nào đó (tìm theo tên để xoá)\n3. Kiểm tra xem giỏ hàng có rỗng không\n4. Tính tổng tiền giỏ hàng\n5. Hiển thị toàn bộ sản phẩm trong giỏiện\n6.Xóa toàn bộ giỏ hàng \n'))
        if k == 1 : self.them_sp()
        if k == 2 : self.xoa_sp()
        if k == 3 : self.ktra()
        if k == 4 : self.tong_tien()
        if k == 5 : self.hien_thi()
        if k == 6 : self.xoa()
        t = input("Bạn có muốn thoát không :")
        if t != 'yes' : self.menu()
        


p = ShoppingCart()
p.menu()
```

```python
# W8A6

class  Rectangle() :
    def __init__ (self,w,h) :
        self.w = w
        self.h = h
    def area(self) :
        self.S = self.w * self.h
        return self.S
    def perimeter(self) :
        self.C = 2*self.w + 2*self.h
        return self.C
    def scale(self,k) :
        self.S = self.S * k**2
        self.C = self.C * k
    def __str__(self) :
        return f'{self.S :.2f} {self.C :.2f}'

w,h,k = list(map(float,input().split()))
p = Rectangle(w,h)
p.area()
p.perimeter()
p.scale(k)
print (p)
```

```python
# W8A7

class Faction() :
    def __init__ (self ,num,den) :
        self.num = num
        self.den = den
        ucln = min(abs(num), abs(den))
        for i in range(ucln,0,-1) :
            if num % i == 0 and den % i == 0 :
                self.num = num // i
                self.den = den // i
                break
    def add(self,other) :
        tu = self.num * other.den + other.num * self.den
        mau = self.den * other.den
        return Faction(tu,mau)
    def sub(self,other) :
        tu = self.num * other.den - other.num * self.den
        mau = self.den * other.den
        return Faction(tu,mau)
    def mul(self,other) :
        tu = self.num * other.num
        mau = self.den * other.den
        return Faction(tu,mau)
    def div(self,other) :
        tu = self.num * other.den
        mau = self.den * other.num
        return Faction(tu,mau)
    def __str__(self) :
        return f'{self.num}/{self.den}'
    
a,b,op,c,d = list(input().split())
a,b,c,d = map(int,[a,b,c,d])
p1 = Faction(a,b)
p2 = Faction(c,d)
if op == '+' : print(p1.add(p2))
if op == '-' : print(p1.sub(p2))    
if op == '*' : print(p1.mul(p2))
if op == '/' : print(p1.div(p2))
    
```

```python
# W8A8

class BankAccount() :
    def __init__(self,owner,balance) :
        self.owner = owner
        self.balance = balance
    def deposit(self,x) :
        self.balance += x
    def withdraw(self,x) :
        if x <=  self.balance :
            self.balance -= x
    def __str__(self) :
        return f"{self.balance :.2f}"

owner , balance = input().split()
balance = float(balance)

p = BankAccount(owner,balance)
    
n = int(input())

for i in range(n) :
    op , x = input().split()
    x = float(x)
    if op == "DEPOSIT" :
        p.deposit(x)
    elif op == "WITHDRAW" :
        p.withdraw(x)
print(p)
```

```python
# W8A9

class Student() :
    def __init__(self,name) :
        self.name = name
        self.scores = {}
    def add_score(self,subject,score) :
        self.scores[subject] = int(score)
    def avg(self) :
        return sum(self.scores.values()) / len(self.scores)
    def rank(self) :
        a = self.avg()
        if a >= 8 : return f'{self.name} {a :.2f} Excellent'
        elif a >= 6.5 and a < 8 : return f'{self.name} {a :.2f} Good'
        elif a >= 5 and a < 6.5 : return f'{self.name} {a :.2f} Average'
        else : return f'{self.name} {a :.2f} Poor'

name = input()
n = int(input())
p = Student(name)
for i in range(n) :
    subject,score = input().split()
    p.add_score(subject,score)
    
print(p.rank())

```

```python
# W8A10

class Book() :
    def __init__(self,title,author,year) :
        self.title = title
        self.author = author 
        self.year = year
    
class Library() :
    def __init__(self) :
        self.t = []
        self.a = []
        self.y = []
    def add(self , book) :
        self.t.append(book.title)
        self.a.append(book.author)
        self.y.append(book.year)
    def count_by_author(self,name) :
        c = 0
        for i in self.a :
            if i == name : c += 1
        return c
    def count_by_year(self , year) :
        c = 0
        for i in self.y :
            if i == year : c += 1
        return c
    
n = int(input())

p1 = Library()
for i in range(n-1) :
    op , b = input().split()
    t , a , y = b.split(';')
    book = Book(t,a,y)
    p1.add(book)
op , x = input().split()
if op == 'COUNT' : print (p1.count_by_author(x))
else : print(p1.count_by_year(x))
```

```python
# W8A11

class Toa_do() :
    def __init__(self ,x,y,z) :
        self.x = x
        self.y = y  
        self.z = z
    def vector(self,other) :
        vx = other.x - self.x
        vy = other.y - self.y   
        vz = other.z - self.z
        return Toa_do(vx,vy,vz)
    def tich_co_huong(self,other) :
        vx = self.y * other.z - self.z * other.y
        vy = self.z * other.x - self.x * other.z
        vz = self.x * other.y - self.y * other.x
        return Toa_do(vx,vy,vz)
    def goc_2vector(self,other) :
        import math
        dot = abs(self.x * other.x + self.y * other.y + self.z * other.z)
        mod1 = math.sqrt(self.x**2 + self.y**2 + self.z**2)
        mod2 = math.sqrt(other.x**2 + other.y**2 + other.z**2)
        cos_theta = dot / (mod1 * mod2)
        angle_rad = math.acos(cos_theta)
        angle_deg = math.degrees(angle_rad)
        return f'{angle_deg :.2f}'

A = Toa_do(*map(float,input().split()))
B = Toa_do(*map(float,input().split()))
C = Toa_do(*map(float,input().split()))
D = Toa_do(*map(float,input().split()))
AB = A.vector(B)
AC = A.vector(C)
DB = D.vector(B)
DC = D.vector(C)
ABC_vtor = AB.tich_co_huong(AC)
BCD_vtor = DB.tich_co_huong(DC)

print( ABC_vtor.goc_2vector(BCD_vtor)) 
```

```python
# W8A12

class ComplexNumber() :
    def __init__(self,re,im) :
        self.re = re
        self.im = im
    def ComplexNumber_multiply(self,other) :
        re = self.re*other.re - self.im*other.im
        im = self.re*other.im + self.im*other.re
        return ComplexNumber(re,im)
    def __str__(self) :
        return f'{self.re} {self.im}'
    
a , b = map(float,input().split())
c , d = map(float,input().split())
p1 = ComplexNumber(a,b)
p2 = ComplexNumber(c,d)

print(p1.ComplexNumber_multiply(p2))
```

```python
# W8A13

class Triangle() :
    def __init__(self,edge1,edge2,edge3) :
        self.edge1 = edge1
        self.edge2 = edge2
        self.edge3 = edge3
    def getArea(self) :
        a , b , c = self.edge1 , self.edge2 , self.edge3
        if a + b > c and a + c > b and b + c > a and a > 0 and b > 0 and c > 0 :
            p = (a + b + c)/2
            area = (p*(p - a)*(p - b)*(p - c))**(0.5)
            return f'{area :.2f}'
        else : return -1

a,b,c = map(int,input().split())
p1 = Triangle(a,b,c)

ans = p1.getArea()
if ans == -1 : print( 'invalid')
else : print (ans)
```
