```python
# W4A8

n = int(input())

buoc = 0
pal = n

def so_doi_xung(a):
    A = str(a)
    B = A[::-1]
    if A in B :
        return 1
    else :
        return -1

def reverse(a) :
    A = str(a)
    B = A[::-1]
    return int(B)

while so_doi_xung(pal) != 1 :
    n += reverse(n)
    buoc += 1
    pal = n
print( buoc,pal)

        
```

```python
# W4A9

n = int(input())

a = int(n**0.5)

def lap(b):
    string = str(b)
    A = []
    for i in string:
        A.append(i)

    for i in A :
        b = -1
        for k in range(0,len(A)) :
            if A[k] == i:
                b += 1
        if b>=1:
            return 1

for i in range(1,a+1):
    m = i**2
    if lap(m) != 1 :
        print(m,end=" ")
    
    

```

```python
# W4A10

def collatz(x) : 
    a = 1
    while x != 1:
        if x % 2== 0 :
            x /= 2
        else :
            x = 3*x +1
        a += 1
    return a

collatz(9)
n = int(input())

max_col = collatz(1)
col = 1

for i in range(1,n + 1) :
    if max_col < collatz(i) :
        col = i
        max_col = collatz(i)
    elif max_col == collatz(i) and i < col :
        col = i

print(col,max_col)

```

```python
# W4A26 

def fib(n) :
    if n == 1 or n == 2 :
        return 1
    else :
        fibo = fib(n-1) + fib(n-2)
        return fibo 
    
A = int(input())
n = 1
while fib(n) <= A :
    n += 1

print(fib(n-1))
```

```python
# W4A32

pass_word = input()

if len(pass_word) <= 6:
    print('mật khẩu yếu ')
else :
    in_hoa = 0
    in_thuong = 0
    so = 0
    dac_biet = 0

    for i in pass_word :
        if 65 <= ord(i) <= 90 :
            in_hoa += 1
        elif 97 <= ord(i) <= 122 :
            in_thuong += 1
        elif 48 <= ord(i) <= 57 :
            so += 1
        elif 35 <= ord(i) <= 47 or 58 <= ord(i) <= 64 :
            dac_biet += 1

    if in_hoa >= 1 and in_thuong >= 1 and so >= 1 and dac_biet >= 1 :
        print('mật khẩu mạnh')
    else :
        print('mật khẩu yếu ')
    
```
