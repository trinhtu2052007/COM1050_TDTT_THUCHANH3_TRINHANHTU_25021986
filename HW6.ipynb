```python
# W6A9

def dictionary(A) :
    C=['{','}',':',"'",' ',',']
    B=[]
    for i in A :
        if i not in C:
            B.append(i)
    
    d_b = {}
    for i in range(0,len(B)-1,2) :
        d_b[B[i]]=int(B[i+1])
    return d_b

A = dictionary(input())
B = dictionary(input())

AB = {}
for key in (A) :
    if key in B :
        AB[key] = A[key] + B[key]
    else :
        AB[key] = A[key]

for key in (B) :
    if key not in A :
        AB[key] = B[key]

print(AB)

```

```python
# W6A10

A = list(map(int,input().split()))
k = int(input())

B = []
index = []

for i in range(len(A)) :
    if k - A[i] in A : # and  i not in index :
        index.append(i)
        for j in range(len(A)) :
            if k - A[i] == A[j] and j not in index and (A[i],A[j]) not in B :
                index.append(j)
                B.append((A[i],A[j]))

B.sort()

print(B)


```

```python
# W6A12

A = list(map(int,input().split()))
var_appear = {}

for i in A :
    if i not in var_appear :
        var_appear[i] = 1
    else :
        var_appear[i] += 1

var = A[0]
max_v = var_appear[A[0]]


for k,v in var_appear.items() :
    if v > max_v :
        var = int(k)
        max_v = v
    if v == max_v and var > int(k) :
        var = int(k)
        max_v = v

print(var)

    
```

```python
# W6A20

def find_pl_var(my_list,var) :
    for i in range(len(my_list)) :
        if my_list[i] == var : return i +1 
    else : return -1

A = list(map(int,input().split())) 
k = int(input())

print(find_pl_var(A,k))
```
